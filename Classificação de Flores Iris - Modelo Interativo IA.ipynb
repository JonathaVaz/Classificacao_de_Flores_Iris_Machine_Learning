```python
from sklearn import datasets
from sklearn import svm
import pandas as pd
import numpy as np
import joblib
```

### **Importando o Tkinter para a Criação do Visual e Leitura de Imagens**

```python
import tkinter
from PIL import ImageTk, Image
```

### **Definindo o Tipo das Flores em cada Variável**

```python
flor_setosa = Image.open("flor_setosa.jpg")
flor_versicolor = Image.open("flor_versicolour.jpg")
flor_virginica = Image.open("flor_virginica.jpg")
```

### **Criação do Modelo Visual Adepto para Usuários**

```python
janela = tkinter.Tk()
janela.geometry('800x600')

flor_setosa = ImageTk.PhotoImage(Image.open("flor_setosa.jpg"))
flor_versicolor = ImageTk.PhotoImage(Image.open("flor_versicolour.jpg"))
flor_virginica = ImageTk.PhotoImage(Image.open("flor_virginica.jpg"))

rotulo_sepala_a = tkinter.Label(janela, text='Digite a altura da Sepala')
rotulo_sepala_a.pack()

sepala_a = tkinter.Entry(janela)
sepala_a.pack()


rotulo_sepala_l = tkinter.Label(janela, text='Digite a largura da Sepala')
rotulo_sepala_l.pack()

sepala_l = tkinter.Entry(janela)
sepala_l.pack()


rotulo_petala_a = tkinter.Label(janela, text='Digite a altura da Petala')
rotulo_petala_a.pack()

petala_a = tkinter.Entry(janela)
petala_a.pack()


rotulo_sepala_l = tkinter.Label(janela, text='Digite a largura da Petala')
rotulo_sepala_l.pack()

petala_l = tkinter.Entry(janela)
petala_l.pack()


resultado = tkinter.Label(janela, text='')
resultado.pack()

def classificar():
    
    sl = float(sepala_l.get())
    sa = float(sepala_a.get())
    pl = float(petala_l.get())
    pa = float(petala_a.get())
    dados = np.array([[sa, sl, pa, pl]])
    botanico = joblib.load('modelo.pkl')
    classificacao = botanico.predict(dados)
    
    
    
    if classificacao == 0:
        resultado['text'] = 'Iris-setosa'
        resultado['image'] = flor_setosa
        
    elif classificacao == 1:
        resultado['text'] = 'Iris-Versicolour'
        resultado['image'] = flor_versicolor
        
        
    else:
        resultado['text'] = 'Iris-Virginica'
        resultado['image'] = flor_virginica

        
botao = tkinter.Button(janela, text='Que Flor é essa?', command=classificar)
botao.pack()

janela.mainloop()
```
